```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import cmasher as cmr
from scipy.ndimage import gaussian_filter

plt.rcParams['axes.labelsize'] = 25
plt.rcParams['axes.titlesize'] = 25
plt.rcParams['xtick.labelsize'] = 25
plt.rcParams['ytick.labelsize'] = 25
plt.rcParams['figure.figsize'] = (11, 10)
```

```python
xrange = np.array([-2.2, 2.2])
yrange = np.array([-2.2, 2.2])
n = 40

x_pos = np.linspace(*xrange, n)
y_pos = np.linspace(*yrange, n)
x_grid, y_grid = np.meshgrid(x_pos, y_pos)
x_direct = np.random.normal(loc=2, size=x_grid.shape)
y_direct = np.random.normal(loc=0, size=y_grid.shape)
color = np.random.normal(x_grid) + np.random.normal(y_grid, 2)
color = gaussian_filter(color, 1)
speed = np.sqrt(x_direct**2 + y_direct**2)
```

```python
plt.axis('off')
plt.xlim(0.9 * xrange)
plt.ylim(0.95 * yrange)
plt.streamplot(x_grid, y_grid, x_direct, y_direct, density=[1, 6], color=color, cmap=cmr.ocean, integration_direction='both', arrowsize=0, minlength=0.3, linewidth=5 * speed / speed.max())
```
